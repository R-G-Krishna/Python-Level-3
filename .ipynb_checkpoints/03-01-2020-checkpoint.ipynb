## How To Use Group By Method In DataFrame

```python
import numpy as np
import pandas as pd
```

```python
df=pd.DataFrame({'A':['Foot','Bar','Foo','Bar','Foo','Bar','Foo','Foo'],
                'B':['One','One','Two','Three','Two','Two','One','Three'],
                'C':np.random.randint(8),'D':np.random.randint(8)})
df
```

```python
df.describe()
```

```python
df.groupby('A').sum()
```

```python
df.groupby(['A','B']).sum()
```

## How To Use Rank Method In DataFrame

```python
df1=pd.DataFrame(data={'Students':['Krishna','Gopi','Kiran','Sai','Vimal','Arjun','Sampath','Jayanth'],
                      'Marks':[89,89,56,76,83,45,45,np.nan]},dtype=int)
df1
```

- default_rank: This is the default behaviour obtained without using any parameter
- max_rank:
- pct_rank:
- NA_bottom:
- NA_top:

```python
df1['default_rank']= df1['Marks'].rank()
df1
```

```python
df1['max_rank']= df1['Marks'].rank(method='max')
df1
```

```python
df1['max_rank']= df1['Marks'].rank(method='min')
df1
```

```python
df1['min_rank']= df1['Marks'].rank(method='max')
df1
```

```python
df1['min_rank']= df1['Marks'].rank(method='min')
df1
```

```python
df1['pct_rank']= df1['Marks'].rank(pct=True)
df1
```

```python
df1['NA_bottom']= df1['Marks'].rank(na_option='top')
df1
```

```python
df1['NA_bottom']= df1['Marks'].rank(na_option='bottom')
df1
```

```python
df1['NA_top']= df1['Marks'].rank(na_option='bottom')
df1
```

```python
df1['NA_top']= df1['Marks'].rank(na_option='top')
df1
```
