```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ds=pd.read_csv("iris.csv")
ds
```

```python
ds['Species'].value_counts()
```

```python
a=ds['SepalLengthCm'].max()
b=ds['SepalWidthCm'].max()
c=ds['PetalLengthCm'].max()
d=ds['PetalWidthCm'].max()
print("max of SepalLength:",a,"\n","max of SepalWidth:",b,"\n","max of PetalLength:",c,"\n","max of PetalWidth:",d)
```

```python
e=ds['SepalLengthCm'].mean()
f=ds['SepalWidthCm'].mean()
g=ds['PetalLengthCm'].mean()
h=ds['PetalWidthCm'].mean()
print("mean of SepalLength:",int(e),"\n","mean of SepalWidth:",int(f),"\n","mean of PetalLength:",int(g),"\n","mean of PetalWidth:",int(h))
```

### Bar Graph

```python
i=ds['SepalLengthCm'].value_counts()
plt.xlabel("SepalLength")
plt.ylabel("PetalLength")
plt.title("SepalLength With PetalLength")
i.sort_index().plot.bar()
plt.show()
```

### Scatter plot

```python
i=ds['SepalLengthCm']
j=ds['PetalLengthCm']
plt.scatter(i,j,label='scatter plot')
plt.xlabel('SepalLength')
plt.ylabel('petalLength')
plt.legend()
plt.show()
```

### Joint Plot

```python
print(sns.jointplot(ds['SepalLengthCm'],ds['SepalWidthCm']))
```
