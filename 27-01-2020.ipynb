```python
import csv
x = []
y = []
with open('krishna.txt','r')as csvfile:
    plots = csv.reader(csvfile,delimiter=',')
    for row in plots:
        x.append(int(row[0]))
        y.append(int(row[1]))
plt.plot(x,y,color='g',label='Loaded from file')
plt.scatter(x,y,label='scatter plot')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
