```python
import matplotlib.pyplot as plt
print("B\n",np.linspace(2.0,3.0,num=5),"\n")
x=np.linspace(0,2,10)
print("A\n",np.sin(x))
def sinplot(flip=1):
    x=np.linspace(0,14,100)
    for i in range(1,8):
        plt.plot(x,np.sin(x+i*.5)*(7-i)*flip)
sinplot()
plt.show()
```
